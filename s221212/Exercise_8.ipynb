### Part 1 a)

We can estimate p using bootstrap by approximating the unknown population with the observed sample. In order to do this, we treat the data as the empirical distribution and then repeatedly draw bootstrap samples of size n with replacement. For each resample we then compute the bootstrap mean and compare its deviation from the original sample mean. Since $\mu$ is unknown, we approximate $\bar{X} - \mu$ by $\bar{X}^* - \bar{X}$, and the estimate of p is then the proportion of bootstrap samples where $a<\bar{X}^* - \bar{X}<b$.

### Part 1 b)

```python
import numpy as np

# Data
x = np.array([56, 101, 78, 67, 93, 87, 64, 72, 80, 69])

n = len(x)
B = 10000  # The number of bootstrap samples

# Original sample mean
x_bar = np.mean(x)

count = 0

for _ in range(B):
    # Bootstrap sample (resample with replacement)
    x_star = np.random.choice(x, size=n, replace=True)
    
    # Bootstrap mean
    x_star_bar = np.mean(x_star)
    
    # Check condition: -5 < (X*bar - Xbar) < 5
    if -5 < (x_star_bar - x_bar) < 5:
        count += 1

p_hat = count / B

print("Bootstrap estimate of p:", p_hat)
```

### Part 2

```python
# Data
x = np.array([5, 4, 9, 6, 21, 17, 11, 20, 7, 10, 21, 15, 13, 16, 8])

n = len(x)
B = 10000  # The number of bootstrap samples

boot_s2 = []

for _ in range(B):
    x_star = np.random.choice(x, size=n, replace=True)
    s2_star = np.var(x_star, ddof=1)
    boot_s2.append(s2_star)

boot_s2 = np.array(boot_s2)

# Bootstrap estimate
var_S2_bootstrap = np.var(boot_s2, ddof=1)

print("Bootstrap estimate of Var(S^2):", var_S2_bootstrap)
```

### Part 3

```python
# Generate Pareto sample
n = 200
shape = 1.05   # Pareto parameter k

x = np.random.pareto(shape, n) + 1  # Standard Pareto(1.05)

# Sample statistics
sample_mean = np.mean(x)
sample_median = np.median(x)

print("Sample mean:", sample_mean)
print("Sample median:", sample_median)

# Bootstrap function
def bootstrap_var(x, stat_func, B=100):
    n = len(x)
    stats = []

    for _ in range(B):
        x_star = np.random.choice(x, size=n, replace=True)
        stats.append(stat_func(x_star))

    stats = np.array(stats)

    return np.var(stats, ddof=1)

# Results
B = 100 # 100 bootstrap replicas

var_mean = bootstrap_var(x, np.mean, B=B)
var_median = bootstrap_var(x, np.median, B=B)

print("Bootstrap Var(mean):", var_mean)
print("Bootstrap Var(median):", var_median)
```

The result changes for each run through, but generally the sample mean is a lot larger than the sample median, which is due to the heavy right tail of the Pareto distribution. The bootstrap variance of the mean is also a lot larger than the bootstrap variance of the median, which indicates that the mean is highly sensitive to extreme observations. The median, on the other hand, is a lot more robust and remains stable across bootstrap samples, as seen by the small variance. This all shows that for heavy-tailed distributions, the median is a more reliable estimator of the central tendency than the mean.
